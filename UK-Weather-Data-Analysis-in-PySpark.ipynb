```python
climate = spark.read.csv('/content/weather.csv', inferSchema=True, header=True)
#Checking out the dataframe to display  number of rows x columns
print("Dataframe Shape: ", (climate.count(), len(climate.columns)))
```

```python
#Determining missing values count in all columns
from pyspark.sql.functions import when, count, col, isnan
climate.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in climate.columns]).show()
```

```python
#Calculating the percentages of values in months
from pyspark.sql.functions import col

climate.groupBy("month") \
  .count() \
  .withColumn('Percentage', (col('count') / climate.count()) * 100 ) \
  .show()
```

```python
#Finding average rainfall with respect to month.November is the highest followed by December. 
climate.groupby("month").agg({"rain": "mean"}).show()
```

```python

# Creating a 'TempView'relational table.
climate.createOrReplaceTempView('Climatechange')
```

```python
#Selecting all the stations
query = spark.sql("Select  *  from Climatechange") 
query.show()
```

```python
# Select all records in Heathrow with temperature >20 and rainfalls >100
query = spark.sql('''
    Select  *  from Climatechange 
    where station='heathrow' 
    and tmax > 20
    and rain > 100
    ''') 
query.show()
```

```python
#Select all records in cardiff with Air Frost numbers > 7 and minimum temperature < 1
query = spark.sql('''
    Select  *  from Climatechange 
    where station='heathrow' 
    and af > 15
    and tmin < 1
    ''') 
query.show()
```

```python
# Group by station and calculate average sun hours, min temp and air frost 

query = spark.sql('''
Select
station as station,
  avg(sun) as Sunshine
  , max(af) as air_frost
  , min(tmin) as min_temperature
from Climatechange 
group by station
''')
query.show()
```
